```python
x=df
```

```python
#elbow method k=4
from sklearn.cluster import KMeans
model=KMeans(n_clusters=4,init='k-means++',max_iter=300,n_init=10,random_state=0)
model.fit(x)
y_pred=model.predict(x)
```

```python
labels=model.labels_
centroid=model.cluster_centers_
```
